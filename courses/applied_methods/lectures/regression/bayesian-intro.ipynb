## Introduction to Bayesian Analysis

In class, we said that Bayesian analysis has a general setup:
- The posterior, $P(X | Data)$
- The likelihood, $P(Data | X)$
- The prior, $P(X)$

For the question of "We tossed a coin, it showed Tails, what is its chance to show Heads?"

We first write down our variable of interest:

P(show Heads | one Tail in one toss)

Let's just rewrite "chance to show Heads" using the variable $p$ and the outcome of the coin toss as a **random** variable $Y$. This allows us to rewrite the posterior as:

$P(p | Y = Tails)$

This would imply that the likelihood is:

$P(Y | p)$

If you recognize this as a Bernoulli random variable, this likelihood is then (assuming $Y=1 \Rightarrow Y=Heads$) 

$P(Y|p) = p^Y(1-p)^{(1-Y)}$

Then we need to decide on the prior, $P(p)$, what do you think $p$ would be before seeing any data?
Below we plot a few density plots using the flexible Beta distribution which takes in 2 parameters.
Which one do you "believe" in?

```r
x <- seq(0, 1, length.out=100)
par(mfrow=c(2, 2))
plot(x, dbeta(x, 1, 1), type="l", main="alpha=1, beta=1")
plot(x, dbeta(x, 2, 2), type="l", main="alpha=2, beta=2")
plot(x, dbeta(x, 1, 3), type="l", main="alpha=1, beta=3")
plot(x, dbeta(x, 3, 1), type="l", main="alpha=3, beta=1")
```

There's no right/wrong answer for the choice of the prior but someone may ask you to explain what shaped your belief.

What is worth noting is that higher alpha values would suggest more belief towards "higher chacne to show Heads".
